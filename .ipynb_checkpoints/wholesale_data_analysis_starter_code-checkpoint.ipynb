## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty

# Create a column for the subtotal for each line
df['subtotal'] = df['unit_price'] * df['qty']

# Display the DataFrame with the new 'subtotal' column
print(df)
      
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

# Define a function to calculate shipping price based on weight
def calculate_shipping_price(weight):
    if weight > 50:
        return 7 * weight
    else:
        return 10 * weight

# Create the 'shipping_price' column by applying the function to 'unit_weight'
df['shipping_price'] = df['unit_weight'].apply(lambda x: x * 7 if x > 50 else x * 10)

# Display the DataFrame with the new 'shipping_price' column
print(df)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%

# Calculate the total price including shipping and sales tax
df['total_price'] = df['subtotal'] + df['shipping_price'] + (df['subtotal'] + df['shipping_price']) * 0.0925

# Display the DataFrame with the new columns
print(df)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

# Create a column for the total cost of each line
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']

# Display the DataFrame to verify the new columns
print(df)
```

```python
# Create a column for the profit of each line using line cost and line price

# Calculate the profit for each line
df['line_cost'] = df['unit_cost'] * df['qty'] + df['unit_weight'] * df['qty']
df['line_price'] = df['unit_price'] * df['qty']
df['profit'] = df['line_price'] - df['line_cost']

# Display the DataFrame with the new 'profit' column
print(df)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

# Calculate total revenue for each client
df['total_revenue'] = df['unit_price'] * df['qty']

# Group by client_id and sum the total revenue
total_revenue_by_client = df.groupby('client_id')['total_revenue'].sum().reset_index()

# Sort by total revenue in descending order
total_revenue_by_client = total_revenue_by_client.sort_values(by='total_revenue', ascending=False)

# Select the top 5 clients
top_5_clients = total_revenue_by_client.head(5)

print(top_5_clients)
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
# Calculate total revenue, shipping price, and profit for each order
df['total_revenue'] = df['unit_price'] * df['qty']
df['total_shipping_price'] = df['unit_cost'] * df['qty']
df['total_profit'] = df['total_revenue'] - df['total_shipping_price']

# Group by client_id and sum the totals
summary_df = df.groupby('client_id').agg(
    total_units_purchased=('qty', 'sum'),
    total_shipping_price=('total_shipping_price', 'sum'),
    total_revenue=('total_revenue', 'sum'),
    total_profit=('total_profit', 'sum')
).reset_index()

# Sort by total revenue in descending order
summary_df = summary_df.sort_values(by='total_revenue', ascending=False)

# Select the top 5 clients
top_5_clients_summary = summary_df.head(5)

print(top_5_clients_summary)

```

```python
# Format the data and rename the columns to names suitable for presentation.


# Rename columns to suitable names
df = df.rename(columns={
    'unit_price': 'revenue',
    'unit_cost': 'cost',
    'qty': 'units_purchased',
    'unit_weight': 'shipping_price'
})

# Define money columns
money_columns = ['revenue', 'cost', 'shipping_price']

# Define function to convert dollar amount to millions
def currency_format_millions(amount):
    return amount / 1000000

# Apply the currency_format_millions function to money columns
df[money_columns] = df[money_columns].apply(currency_format_millions)

# Rename columns to reflect the change in the money format
df = df.rename(columns={
    'revenue': 'revenue_millions',
    'cost': 'cost_millions',
    'shipping_price': 'shipping_price_millions'
})

print(df) 


```
